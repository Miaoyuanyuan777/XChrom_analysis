## First, combine different samples within the same species, and then combine different species based on homologous genes

```python
import scanpy as sc
import scanpy.external as sce
import numpy as np
import anndata as ad
import seaborn as sns
import pandas as pd
```

```python
sc.set_figure_params(figsize=(3,3), frameon=False)
sns.set_theme()
%config InlineBackend.print_figure_kwargs={"facecolor": "w"}
%config InlineBackend.figure_format="retina"
```

```python
path = './data/'
homo = pd.read_csv('./orthologs_1to1_human_mouse_macaque_marmoset_8904.csv')
out_dir = './harmony_pca/'
```

```python
s11 = sc.read_h5ad(f'{path}/1_human/m1d1n/ad_rna.h5ad')
s12 = sc.read_h5ad(f'{path}/1_human/m1d1/ad_rna.h5ad')
s13 = sc.read_h5ad(f'{path}/1_human/m1d2n/ad_rna.h5ad')
s14 = sc.read_h5ad(f'{path}/1_human/m1d2/ad_rna.h5ad')
s15 = sc.read_h5ad(f'{path}/1_human/m1d3n/ad_rna.h5ad')

s21 = sc.read_h5ad(f'{path}/2_mouse/mop2c1/ad_rna.h5ad')
s22 = sc.read_h5ad(f'{path}/2_mouse/mop2c2/ad_rna.h5ad')
s23 = sc.read_h5ad(f'{path}/2_mouse/mop3c1/ad_rna.h5ad')
s24 = sc.read_h5ad(f'{path}/2_mouse/mop3c2/ad_rna.h5ad')
s25 = sc.read_h5ad(f'{path}/2_mouse/mop4b1/ad_rna.h5ad')
s26 = sc.read_h5ad(f'{path}/2_mouse/mop4b2/ad_rna.h5ad')
s27 = sc.read_h5ad(f'{path}/2_mouse/mop5d1/ad_rna.h5ad')
s28 = sc.read_h5ad(f'{path}/2_mouse/mop5d2/ad_rna.h5ad')

s31 = sc.read_h5ad(f'{path}/3_macaque/Dheni/ad_rna.h5ad')
s32 = sc.read_h5ad(f'{path}/3_macaque/Mac3/ad_rna.h5ad')
s33 = sc.read_h5ad(f'{path}/3_macaque/R2/ad_rna.h5ad')

s41 = sc.read_h5ad(f'{path}/4_marmoset/Webster/ad_rna.h5ad')
s42 = sc.read_h5ad(f'{path}/4_marmoset/Meadow/ad_rna.h5ad')
s43 = sc.read_h5ad(f'{path}/4_marmoset/Peregrine_rep2/ad_rna.h5ad')
s44 = sc.read_h5ad(f'{path}/4_marmoset/Peregrine_rep1/ad_rna.h5ad')
```

```python
s11.var_names = s11.var['ids']
s12.var_names = s12.var['ids']
s13.var_names = s13.var['ids']
s14.var_names = s14.var['ids']
s15.var_names = s15.var['ids']
s21.var.names = s21.var.names.astype('str')
s22.var.names = s22.var.names.astype('str')
s23.var.names = s23.var.names.astype('str')
s24.var.names = s24.var.names.astype('str')
s25.var.names = s25.var.names.astype('str')
s26.var.names = s26.var.names.astype('str')
s27.var.names = s27.var.names.astype('str')
s28.var.names = s28.var.names.astype('str')
```

## 1-concatenate human samples

```python
s11.var_names_make_unique()
s12.var_names_make_unique()
s13.var_names_make_unique()
s14.var_names_make_unique()
s15.var_names_make_unique()
```

```python
s1 = s11.concatenate(s12, s13,s14,s15,join='inner', batch_key='sampleID')
print('human anndata:\n',s1)
print(s1.obs['sampleID'].value_counts(),s1.obs_names[:2])
s1.obs.index = s1.obs.index.str.replace(r'-\d+$', '', regex=True)
```

## 2-concatenate mouse samples

```python
s21.var_names_make_unique()
s22.var_names_make_unique()
s23.var_names_make_unique()
s24.var_names_make_unique()
s25.var_names_make_unique()
s26.var_names_make_unique()
s27.var_names_make_unique()
s28.var_names_make_unique()
```

```python
s2 = s21.concatenate(s22, s23,s24,s25,s26,s27,s28,join='inner', batch_key='sampleID')
print('mouse anndata:\n',s2)
print(s2.obs['sampleID'].value_counts(),s2.obs_names[:2])
s2.obs.index = s2.obs.index.str.replace(r'-\d+$', '', regex=True)
```

## 3-concatenate macaque samples

```python
s31.var_names_make_unique()
s32.var_names_make_unique()
s33.var_names_make_unique()
```

```python
# 合并每个物种的3个样本
s3 = s31.concatenate(s32, s33,join='inner', batch_key='sampleID')
print('macaque anndata:\n',s3)
print(s3.obs['sampleID'].value_counts(),s3.obs_names[:2])
s3.obs.index = s3.obs.index.str.replace(r'-\d+$', '', regex=True)
```

## 4-concatenate marmoset samples

```python
s41.var_names_make_unique()
s42.var_names_make_unique()
s43.var_names_make_unique()
s44.var_names_make_unique()
```

```python
# 合并每个物种的3个样本
s4 = s41.concatenate(s42, s43,s44,join='inner', batch_key='sampleID')
print('marmoset anndata:\n',s4)
print(s4.obs['sampleID'].value_counts(),s4.obs_names[:2])
s4.obs.index = s4.obs.index.str.replace(r'-\d+$', '', regex=True)
```

## 5-concatenate 4 species

```python
## extract orthologs-human
h_genes = homo['ensembl_gene_id']
s_genes = set(s1.var['ids'].values)
matching_genes = list(set(h_genes).intersection(s_genes))
print(len(matching_genes))
f_homo = homo[homo['ensembl_gene_id'].isin(matching_genes)]
s1sub = s1[:, s1.var['ids'].isin(matching_genes)]
# Assign the ensembl_id of humans to s1
subs1_var_ids = s1sub.var['ids'].values
f_homo = f_homo.set_index('ensembl_gene_id')
f_homo_sorted = f_homo.reindex(subs1_var_ids)
f_homo_sorted = f_homo_sorted.reset_index()
s1sub.var['ortho_h_id'] = f_homo_sorted['ensembl_gene_id'].values
s1sub.var['ortho_h_names'] = f_homo_sorted['external_gene_name'].values
```

```python
## extract orthologs-mouse
h_genes = homo['mmusculus_homolog_ensembl_gene']
s_genes = set(s2.var['ids'].values)
matching_genes = list(set(h_genes).intersection(s_genes))
print(len(matching_genes))
f_homo = homo[h_genes.isin(matching_genes)]
s2sub = s2[:, s2.var['ids'].isin(matching_genes)]
# Assign the ensembl_id of humans to s2
subs2_var_ids = s2sub.var['ids'].values
f_homo = f_homo.set_index('mmusculus_homolog_ensembl_gene')
f_homo_sorted = f_homo.reindex(subs2_var_ids)
f_homo_sorted = f_homo_sorted.reset_index()
s2sub.var['ortho_h_id'] = f_homo_sorted['ensembl_gene_id'].values
s2sub.var['ortho_h_names'] = f_homo_sorted['external_gene_name'].values
```

```python
## extract orthologs-macaque
a = 'mmulatta_homolog_associated_gene_name'
h_genes = homo[a]
s_genes = set(s3.var['names'].values)
matching_genes = list(set(h_genes).intersection(s_genes))
print(len(matching_genes))
f_homo = homo[homo[a].isin(matching_genes)]
s3sub = s3[:, s3.var['names'].isin(matching_genes)]
# Assign the ensembl_id of humans to s3
subs3_var_ids = s3sub.var['names'].values
f_homo = f_homo.set_index(a)
f_homo_sorted = f_homo.reindex(subs3_var_ids)
f_homo_sorted = f_homo_sorted.reset_index()
s3sub.var['ortho_h_id'] = f_homo_sorted['ensembl_gene_id'].values
s3sub.var['ortho_h_names'] = f_homo_sorted['external_gene_name'].values
```

```python
## extract orthologs-marmoset
a = 'cjacchus_homolog_associated_gene_name'
h_genes = homo[a]
s_genes = set(s4.var['names'].values)
matching_genes = list(set(h_genes).intersection(s_genes))
print(len(matching_genes))
f_homo = homo[homo[a].isin(matching_genes)]
s4sub = s4[:, s4.var['names'].isin(matching_genes)]
# Assign the ensembl_id of humans to s4
subs4_var_ids = s4sub.var['names'].values
f_homo = f_homo.set_index(a)
f_homo_sorted = f_homo.reindex(subs4_var_ids)
f_homo_sorted = f_homo_sorted.reset_index()
s4sub.var['ortho_h_id'] = f_homo_sorted['ensembl_gene_id'].values
s4sub.var['ortho_h_names'] = f_homo_sorted['external_gene_name'].values
```

```python
s1sub.var_names = s1sub.var['ortho_h_id']
s2sub.var_names = s2sub.var['ortho_h_id']
s3sub.var_names = s3sub.var['ortho_h_id']
s4sub.var_names = s4sub.var['ortho_h_id']
combined = s1sub.concatenate(s2sub, s3sub, s4sub, join='inner', batch_key='species')
print(combined.shape)
```

```python
combined = s1sub.concatenate(s2sub, s3sub, s4sub, join='inner', batch_key='species')
print(combined)
combined.obs['batch'] = combined.obs['species'].astype(str) + '_' + combined.obs['sampleID'].astype(str)
print('species_samples:\n',combined.obs['batch'].value_counts())
```

```python
adata = combined.copy()
adata.write_h5ad(f'{out_dir}/beforeBatch_combined4species_rna.h5ad')
```

```python
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata)
adata = adata[:, adata.var.highly_variable]
sc.pp.scale(adata, max_value=10)
sc.pp.pca(adata, n_comps=32)
```

```python
# 基于校正前的 PCA 空间计算邻居
sc.pp.neighbors(adata, use_rep='X_pca')
# UMAP 降维
sc.tl.umap(adata)
# 绘制 UMAP 图，按批次和细胞类型上色
sc.pl.umap(adata, color=['batch','cell_type'])
sc.pl.umap(adata, color=['species','sampleID'])
```

```python
sce.pp.harmony_integrate(adata, 'batch', max_iter_harmony=30)
'X_pca_harmony' in adata.obsm
```

```python
# 基于校正后的 PCA 空间计算邻居
sc.pp.neighbors(adata, use_rep='X_pca_harmony')
# UMAP 降维
sc.tl.umap(adata)
# 绘制 UMAP 图，按批次和细胞类型上色
sc.pl.umap(adata, color=['batch','cell_type'])
sc.pl.umap(adata, color=['species','sampleID'])
```

```python
adata.obs.index = adata.obs.index.str.replace(r'-\d+$', '', regex=True)
adata.write_h5ad(f'{out_dir}/batch10C_rna.h5ad')
```
