```python
setup_seed(seed)
# scBasset/scbasset/utils.py
'''将0123的seq编码转换为one hot， 以及  将标签转为二值信号(0/1)'''
class generator:
    def __init__(self,seq,peakid,cellembed,b,m):
        self.file = h5py.File(seq, 'r') 
        self.data = self.file['X'][peakid]
        self.m = m
        self.n_cells = m.shape[1]
        self.cellembed = cellembed
        self.b = b
        self.ones = np.ones(1344)
        self.rows = np.arange(1344)

    def __call__(self):
        for i in range(self.data.shape[0]):  ## peaks
            x1 = self.data[i]  ## 得到的x1是包含0，1，2，3数字元素的list
            x_tf = sparse.coo_matrix((self.ones, (self.rows, x1)),   
            ## 这里构建了一个矩阵，根据行元素self.rows和列元素x1填充1，如[0,3]即在第0行第3列填充1，因此最终输出是max(self.rows)行，max(x1)列
                                               shape=(1344, 4), 
                                               dtype='int8').toarray()
            y = self.m.indices[self.m.indptr[i]:self.m.indptr[i+1]]  ##indices需要提取细胞标签
            y_tf = np.zeros(self.n_cells, dtype='int8')
            y_tf[y] = 1
            x2 = self.cellembed
            b = self.b
            yield (x_tf,x2,b),y_tf
```

```python
# 噪声注入仅在训练时有效
def add_noise(inputs,noise_factor=0.01):
    return inputs + noise_factor * tf.random.normal(shape=tf.shape(inputs))
    
def Cmodel(n_cells, cell_vec=32, seq_len=1344, show_summary=True, noise_factor=0.01):
    sequence = tf.keras.Input(shape=(seq_len, 4), name="sequence")
    cell_input = tf.keras.Input(shape=(n_cells, cell_vec), name='cell_embed')
    seq_depth = tf.keras.Input(shape=(n_cells), name='sequencing_depth')

    # Your current model structure
    (x1, reverse_bool,) = StochasticReverseComplement()(sequence)
    x1 = StochasticShift(3)(x1)
    x1 = conv_block(x1, filters=288, kernel_size=17, pool_size=3)
    x1 = conv_tower(x1, filters_init=288, filters_mult=1.122, repeat=6, kernel_size=5, pool_size=2)
    x1 = conv_block(x1, filters=256, kernel_size=1)
    x1 = dense_block(x1, flatten=True, units=cell_vec, dropout=0.2)
    x1 = GELU()(x1)
    x1 = tf.squeeze(x1, axis=1)   # (bsz, 32)
    x1 = tf.expand_dims(x1, axis=-1)  # (bsz, 32, 1)

    # Add noise to cell_input
    noisy_cell_input = tf.keras.layers.Lambda(lambda x: tf.keras.backend.in_train_phase(
        add_noise(x, noise_factor=noise_factor), x))(cell_input)
    x2 = tf.keras.layers.LayerNormalization()(noisy_cell_input)
    x2 = tf.keras.layers.Dense(64, activation='relu')(x2)
    x2 = tf.keras.layers.Dense(32, activation='linear', name='final_cellembed')(x2)

    b = tf.expand_dims(seq_depth, axis=-1)  # (bsz,n_cells,1)
    b = tf.keras.layers.Dense(1, activation='linear')(b)
    b = tf.squeeze(b, axis=-1)  # (bsz,n_cells)

    multiplied = tf.matmul(x2, x1)  # (bsz, num_cells, 1)
    multiplied = tf.squeeze(multiplied, axis=-1)  # Remove the last dimension (bsz, num_cells)

    output = tf.keras.activations.sigmoid(multiplied + b)

    # Model with two outputs: x2 (cell embeddings) and final output
    model = tf.keras.Model(inputs=(sequence, cell_input, seq_depth), outputs=output)

    if show_summary:
        model.summary()
    return model
```

```python
def cluster_metric(ad_rna, ad_atac,n=100,label='pca32_leiden',use_rep_rna = 'X_pca',use_rep_atac='projection'):
    sc.pp.neighbors(ad_rna, n_neighbors=n+1, use_rep=use_rep_rna)
    m_RNA_neighbors = [i.indices for i in ad_rna.obsp['distances']] # neighbor idx for each cell

    # ATAC neighbors
    sc.pp.neighbors(ad_atac, n_neighbors=n+1, use_rep=use_rep_atac) # n nearest neighbor of each cell
    m_ATAC_neighbors = [i.indices for i in ad_atac.obsp['distances']] # neighbor idx for each cell

    # neighbor score
    n_shared_neighbors = np.mean([len(np.intersect1d(i,j)) for i,j in zip(m_RNA_neighbors, m_ATAC_neighbors)])

    # label score
    neighbor_label = ad_atac.obs[label].values[np.concatenate(m_ATAC_neighbors, axis=0)] 
    # label for all neighbors[第0个细胞的neighbors的label 第1个细胞的neighbors的label ... 第len(m_ATAC_neighbors)的.. ]
    cell_label = ad_atac.obs[label].values[np.repeat(np.arange(len(m_ATAC_neighbors)), [len(j) for j in m_ATAC_neighbors])] # label for all cells
    ## 第0个细胞，重复label=2，第0个细胞有几个neighbors就重复多少次2
    ## [第0个细胞的label重复对应neighbors个次 第1个细胞的label重复对应neighbors个次 ...]
    n_shared_labels = (neighbor_label==cell_label).sum() / len(m_ATAC_neighbors)
    ## sum是对所有细胞的情况求和，因此除掉细胞的个数，表示平均每个细胞的情况
    sc.tl.umap(ad_atac) ## 生成UMAP的X_UMAP二维坐标
    return n_shared_neighbors/n, n_shared_labels/n
## 有些细胞可能在距离计算中遇到边界情况，或者与其他细胞距离相同，导致某些细胞的邻居数量与其他细胞不同
```

```python
ad = anndata.read_h5ad(f'{preprocess_folder}/ad.h5ad')  ## 提取细胞
m = sparse.load_npz('%s/m.npz'%preprocess_folder).tocsr()  # 维持为 CSR 格式
all_seq = '%s/all_seqs.h5'%preprocess_folder

cell_embed_test = np.array(ad.obsm['zscore32_perpc'])
b_test = np.array(ad.obs['b_zscore'].values)

idx = sorted(range(ad.shape[1]))

print(cell_embed_test.shape,b_test.shape)
print(cell_embed_test,b_test)
print(m,m.shape)   ##(30752, 2711)
```

```python
m_to1 = m.copy()  # 复制原始稀疏矩阵
m_to1[m_to1 != 0] = 1  # 将非零值设置为1
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score, average_precision_score

# `m_crosscell_to1` 是一个形状为 (n_peaks, n_cells) 的数组，包含二值标签
# `pred` 是一个同样形状的数组，包含对应的预测概率
test_prediction = pred
mt1_dense = m_to1.toarray()


# 将所有细胞的标签和预测展平成一个一维数组
labels_tensor = tf.constant(mt1_dense)  ##(24436, 271)
pred_tensor = tf.constant(test_prediction)  ##(24436, 271)
y_true = tf.reshape(labels_tensor, [-1])  # -1 表示自动推断维度
y_pred = tf.reshape(pred_tensor, [-1])

# 计算整体的 AUC
overall_auc = roc_auc_score(y_true, y_pred)
# 计算整体的 PR
overall_pr = average_precision_score(y_true, y_pred)

print("Overall ROC AUC:", overall_auc)
print("Overall PR AUC:", overall_pr)
```

```python
import math
auc_values = []
pr_values = []
peak_cnts = []

# 将稀疏矩阵转为稠密矩阵一次性完成，而不是每次迭代时都转换
for j in range(test_prediction.shape[1]):  # per cell (24436, 271)
    # 获取列数据
    y_true = mt1_dense[:, j]
    y_pred = test_prediction[:, j]
    # 检查 y_true 中的类别数量
    unique_classes = np.unique(y_true)
    if len(unique_classes) == 1:
        # 如果只有一个类别，跳过计算
        continue
    # 计算 ROC AUC 和 PR
    auc = roc_auc_score(y_true, y_pred)
    pr = average_precision_score(y_true, y_pred)
    # 计算峰值计数
    peak_cnt = np.sum(y_true)
    # 将结果添加到列表中
    auc_values.append(auc)
    pr_values.append(pr)
    peak_cnts.append(peak_cnt)
# 计算平均值
avg_auc = sum(auc_values) / len(auc_values)
avg_pr = sum(pr_values) / len(pr_values)
# 对于峰值计数取对数
log_peak_cnts = [math.log2(cnt) for cnt in peak_cnts if cnt > 0]  # 避免对 0 取对数
print(avg_auc, avg_pr)
```

```python
# 计算相关系数
R_auc_matrix = np.corrcoef(auc_values, log_peak_cnts)
R_pr_matrix = np.corrcoef(pr_values, log_peak_cnts)
R_auc = R_auc_matrix[0, 1]
R_pr = R_pr_matrix[0, 1]

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(auc_values, log_peak_cnts, color='blue', label='auROC',s=20, alpha=0.5)
plt.xlabel('auROC per cell')
plt.ylabel('log_peak_count')
plt.title(f'mean = {avg_auc:.4f}')
plt.text(0.87, 13, f'R: {R_auc:.2f}', fontsize=12, color='blue') ## 将浮点数格式化为带有两位小数的字符串
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(pr_values, log_peak_cnts, color='red', label='auPR',s=20, alpha=0.5)
plt.xlabel('auPR per cell')
plt.ylabel('log_peak_count')
plt.title(f'mean = {avg_pr:.4f}')
plt.legend()
plt.text(0., 13, f'R: {R_pr:.2f}', fontsize=12, color='blue')
plt.savefig(f'{out_dir}/cnt_percell.pdf', format='pdf', dpi=300)
plt.show()
```

```python
setup_seed(seed)  # 假设 setup_seed 已经定义
auc_values = []
pr_values = []
cell_cnts = []
# 遍历每一行（peaks）计算AUC
for k in range(test_prediction.shape[0]):  # per peak (24436, 271)
    # 获取当前行的标签和预测值
    y_true = np.array(mt1_dense[k, :])
    y_pred = test_prediction[k, :]
    # 检查 y_true 中的类别数量，避免只有一个类别的情况
    unique_classes = np.unique(y_true)
    if len(unique_classes) == 1:
        # 如果只有一个类别，跳过该行，不能计算 AUC 和 PR
        continue
    # 计算 AUC 和 PR
    auc = roc_auc_score(y_true, y_pred)
    pr = average_precision_score(y_true, y_pred)

    # 计算细胞计数
    cell_cnt = np.sum(y_true)
    
    # 将结果添加到列表中
    auc_values.append(auc)
    pr_values.append(pr)
    cell_cnts.append(cell_cnt)

# 计算 AUC 和 PR 的平均值
avg_auc = sum(auc_values) / len(auc_values)
avg_pr = sum(pr_values) / len(pr_values)

# 对细胞计数取对数，避免 0 的情况
log_cell_cnts = [math.log2(cnt) for cnt in cell_cnts if cnt > 0]
# 打印结果
print(f"Average AUC: {avg_auc}, Average PR: {avg_pr}")
```

```python
# 计算相关系数
R_auc_matrix = np.corrcoef(auc_values, log_cell_cnts)
R_pr_matrix = np.corrcoef(pr_values, log_cell_cnts)
R_auc = R_auc_matrix[0, 1]
R_pr = R_pr_matrix[0, 1]

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(auc_values, log_cell_cnts, color='blue', label='auROC',s=20, alpha=0.5)
plt.xlabel('auROC per peak')
plt.ylabel('log_cell_count')
plt.title(f'mean = {avg_auc:.4f}')
plt.text(0.41, 10, f'R: {R_auc:.2f}', fontsize=12, color='blue') ## 将浮点数格式化为带有两位小数的字符串
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(pr_values, log_cell_cnts, color='red', label='auPR',s=20, alpha=0.5)
plt.xlabel('auPR per peak')
plt.ylabel('log_cell_count')
plt.title(f'mean = {avg_pr:.4f}')
plt.legend()
plt.text(0.03, 10, f'R: {R_pr:.2f}', fontsize=12, color='blue')

plt.savefig(f'{out_dir}/cnt_perpeak.pdf', format='pdf', dpi=300)
plt.show()
```
