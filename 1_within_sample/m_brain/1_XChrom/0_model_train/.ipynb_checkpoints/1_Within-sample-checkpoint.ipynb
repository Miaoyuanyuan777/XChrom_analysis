## within-sample train,prediction and evaluation
### 1. Download data (mouse brain)
- 原始数据位于：https://support.10xgenomics.com/single-cell-multiome-atac-gex/datasets/2.0.0/e18_mouse_brain_fresh_5k
- 可以通过该链接直接下载h5文件：https://cf.10xgenomics.com/samples/cell-arc/2.0.0/e18_mouse_brain_fresh_5k/e18_mouse_brain_fresh_5k_filtered_feature_bc_matrix.h5
- 然后将下载的h5数据save in './data/1_within_sample/18_mouse_brain_fresh_5k_filtered_feature_bc_matrix.h5'     

### 2.Read and filter data

```python
import xchrom as xc
import scanpy as sc
```

```python
input_file = './data/1_within_sample/e18_mouse_brain_fresh_5k_filtered_feature_bc_matrix.h5'
output_file = './data/1_within_sample/m_brain_paired.h5ad'
ad_rna, ad_atac = xc.read_10x_h5_to_h5ad(
    h5_file=input_file,
    paired=True,
    output_file=output_file
)
```

This will generate three h5ad files: m_brain_paired.h5ad, m_brain_paired_atac.h5ad and m_brain_paired_rna.h5ad in './data/1_within_sample'

```python
### filter data
ad_rna, ad_atac = xc.pp.filter_multiome_data(
    ad_rna = ad_rna,
    ad_atac = ad_atac,
    species = 'mouse'
)
```

### 3. Process scRNA-seq data to generate raw cell embeddings

```python
# ad_rna = sc.read_h5ad('./data/1_within_sample/m_brain_paired_rna.h5ad')
ad_rna = xc.tl.calc_pca(ad_rna, n_comps=32)
sc.pp.neighbors(ad_rna)
print('The dimension of raw cell embeddings: ',ad_rna.obsm["X_pca"].shape)
sc.tl.umap(ad_rna)
sc.tl.leiden(ad_rna, key_added='pc32_leiden', resolution=0.3)
```

Now we have raw cell embedding in ad_rna['X_pca'], and leden clusters in ad_rna['pc32_leiden'] as cell types if needed.

```python
prefix = str(output_file).split('.h5ad')[0]
ad_rna.write(f'{prefix}_rna.h5ad')
ad_atac.write(f'{prefix}_atac.h5ad')
```

### 4. Train/test data prepare

```python
process_folder = './data/1_within_sample/train_data/'
input_fasta = '/picb/bigdata/project/miaoyuanyuan/mm10.fa'
# ad_atac = sc.read_h5ad('./data/1_within_sample/m_brain_paired_atac.h5ad')
dict = xc.pp.process_train_test_single(
    ad_atac=ad_atac, ## can be a str, Path, or anndata.AnnData object
    input_fasta=input_fasta,
    output_path=process_folder
)
```

### 5. Train XChrom

```python
history = xc.tr.train_XChrom(
    input_folder='./data/1_within_sample/train_data/',
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    cellembed_raw='X_pca',
    out_path='./data/1_within_sample/train_out/',
    trackscore = True,
    celltype = 'pc32_leiden',
    epochs = 1000,
    save_freq = 1000,
    verbose = 1
)
```

```python
xc.pl.plot_train_history(
    history = history['history'],
    savefig = True,
    out_file = './data/1_within_sample/train_out/train_history_plot.pdf'
    )
```

### 6. Evaluate the model performance
- on cross-cell,cross-peak,cross-both prediction in within sample scenario

```python
#### 1- calculate cross-cell auROC & auPRC
metrics1 = xc.tl.crosscell_aucprc(
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    input_folder='./data/1_within_sample/train_data',
    model_path='./data/1_within_sample/train_out/E1000best_model.h5',
    output_path='./data/1_within_sample/eval_out',
    cell_input_key='zscore32_perpc',
    save_pred=True,
    scatter_plot=True
    )
```

```python
#### 2- calculate cross-peak auROC & auPRC
metrics2 = xc.tl.crosspeak_aucprc(
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    input_folder='./data/1_within_sample/train_data',
    model_path='./data/1_within_sample/train_out/E1000best_model.h5',
    output_path='./data/1_within_sample/eval_out',
    cell_input_key='zscore32_perpc',
    save_pred=True,
    scatter_plot=True
    )
```

```python
#### 3- calculate crossboth auROC & auPRC
metrics3 = xc.tl.crossboth_aucprc(
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    input_folder='./data/1_within_sample/train_data',
    model_path='./data/1_within_sample/train_out/E1000best_model.h5',
    output_path='./data/1_within_sample/eval_out',
    cell_input_key='zscore32_perpc',
    save_pred=True,
    scatter_plot=True
    )
```

```python
#### 4- calculate crosscell ns & ls
metrics4 = xc.tl.crosscell_nsls(
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    input_folder='./data/1_within_sample/train_data',
    model_path='./data/1_within_sample/train_out/E1000best_model.h5',
    output_path='./data/1_within_sample/eval_out',
    cell_input_key='zscore32_perpc',
    cellembed_raw='X_pca',
    celltype='pc32_leiden',
    save_pred=True,
    plot_umap=True
    )
```

```python
#### 5- calculate denoise ns & ls
metrics5 = xc.tl.denoise_nsls(
    cell_embedding_ad='./data/1_within_sample/m_brain_paired_rna.h5ad',
    input_folder='./data/1_within_sample/train_data',
    model_path='./data/1_within_sample/train_out/E1000best_model.h5',
    output_path='./data/1_within_sample/eval_out',
    cell_input_key='zscore32_perpc',
    cellembed_raw='X_pca',
    celltype='pc32_leiden',
    save_pred=True,
    plot_umap=True
    )
```
