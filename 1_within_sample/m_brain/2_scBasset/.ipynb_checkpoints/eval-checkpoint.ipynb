## Evaluate the training results of the scbasset model--e18
### --Including: the dimensionality reduction clustering effect of the denoising matrix, and the prediction accuracy across regions

```python
import anndata
import tensorflow as tf
import numpy as np
import math
import scanpy as sc
import matplotlib.pyplot as plt
from utils import *
import xchrom as xc
```

```python
path = '../'
traindata_folder = f'{path}/2_scBasset/processed/'
out_dir = f'{path}/2_scBasset/eval_out/'
```

```python
trained_model = f'{path}/2_scBasset/output/best_model.h5'
```

```python
# read h5ad file
ad = anndata.read_h5ad(f'{traindata_folder}/ad.h5ad')
# load model
model = make_model(32, ad.shape[0], show_summary=False)
model.load_weights(trained_model)
```

```python
ad
```

```python
rna = anndata.read_h5ad(f'{path}/0_data/processed_data/m_brain_paired_rna.h5ad')
rna
```

```python
assert ad.obs.index.equals(rna.obs.index), "Indexes differ!"
ad.obs['pc32_leiden'] = rna.obs['pc32_leiden']
```

## denoise results

```python
all_seqs = f'{traindata_folder}/all_seqs.h5' # from the preprocess step
# make tesorflow dataset to feed into model
m = ad.X.tocoo().transpose().tocsr()
n_cells = ad.shape[0]
all_ds = tf.data.Dataset.from_generator(
    generator(all_seqs, m), 
    output_signature=(
            tf.TensorSpec(shape=(1344,4), dtype=tf.int8),
            tf.TensorSpec(shape=(n_cells), dtype=tf.int8),
    )
).batch(128).prefetch(tf.data.AUTOTUNE)
Y_impute = imputation_Y_normalize(all_ds, model)
```

```python
adp = ad.copy()
adp.X=Y_impute.transpose(1,0)
adp.write_h5ad(f'{out_dir}/scBasset_denoise.h5ad')
```

```python
## raw_ATAC
ad1 = ad.copy()
ad1 = xc.tl.calc_pca(ad1)
ns,ls = xc.tl.calc_nsls_score(rna,ad1,100,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(100)={ns:.4f},label score(100)={ls:.4f}')
ns,ls = xc.tl.calc_nsls_score(rna,ad1,50,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(50)={ns:.4f},label score(50)={ls:.4f}')
ns,ls = xc.tl.calc_nsls_score(rna,ad1,10,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(10)={ns:.4f},label score(10)={ls:.4f}')

f, ax= plt.subplots(1,1, figsize=(6, 4))
sc.pl.umap(ad1, color='pc32_leiden', ax=ax,show=False)
plt.tight_layout()
```

```python
ad1 = adp.copy()
ad1 = xc.tl.calc_pca(ad1)
ns,ls = xc.tl.calc_nsls_score(rna,ad1,100,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(100)={ns:.4f},label score(100)={ls:.4f}')
ns,ls = xc.tl.calc_nsls_score(rna,ad1,50,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(50)={ns:.4f},label score(50)={ls:.4f}')
ns,ls = xc.tl.calc_nsls_score(rna,ad1,10,'pc32_leiden',use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(10)={ns:.4f},label score(10)={ls:.4f}')

f, ax= plt.subplots(1,1, figsize=(6, 4))
sc.pl.umap(ad1, color='pc32_leiden', ax=ax,show=False)
plt.tight_layout()
```

## test peaks的预测准确性

```python
test_seqs = f'{traindata_folder}/test_seqs.h5' # from the preprocess step
# make tesorflow dataset to feed into model
test_m = sparse.load_npz(f'{traindata_folder}/m_test.npz')
n_cells = ad.shape[0]
test_ds = tf.data.Dataset.from_generator(
    generator(test_seqs, test_m), 
    output_signature=(
            tf.TensorSpec(shape=(1344,4), dtype=tf.int8),
            tf.TensorSpec(shape=(n_cells), dtype=tf.int8),
    )
).batch(test_m.shape[0]).prefetch(tf.data.AUTOTUNE)
```

```python
for inputs, labels in test_ds:
    pred = model.predict(inputs)
    print(pred.shape)
```

```python
np.save(f'{out_dir}/scBasset_pred_crossRegion.npy', pred)
```

```python
labels = sparse.load_npz(f'{traindata_folder}/m_test.npz').toarray()
labels = (labels >= 1).astype(np.int32)  # It has already been binarized
```

```python
print("Pred shape:", pred.shape)
print("Test labels shape:", labels.shape)
```

```python
labels
```

```python
labels = labels.T
pred = pred.T
```

```python
## -1 Calculate overall auROC & auPRC 
overall_metrics = xc.tl.calc_auc_pr(labels, pred, 'overall')
print(f"Overall auROC: {overall_metrics['auroc']:.4f}, auPRC: {overall_metrics['auprc']:.4f}")

## -2 Calculate per cell auRPC & auPRC
cell_metrics = xc.tl.calc_auc_pr(labels, pred, 'percell')
print(f"Per-cell auROC: {cell_metrics['auroc']:.4f}, auPRC: {cell_metrics['auprc']:.4f}")
print(f"Valid cells: {cell_metrics['n_cells']}")

## -3 Calculate per peak auROC & auPRC
peak_metrics = xc.tl.calc_auc_pr(labels, pred, 'perpeak')
print(f"Per-peak auROC: {peak_metrics['auroc']:.4f}, auPRC: {peak_metrics['auprc']:.4f}")
print(f"Valid peaks: {peak_metrics['n_peaks']}")
```
