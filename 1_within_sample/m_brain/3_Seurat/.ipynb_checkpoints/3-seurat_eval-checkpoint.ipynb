## ⭐ cross-cell prediction
## cross-cell prediciton test(auROC,auPRC)，and cross-cell impute test(ns,ls)

```python
import feather
import scanpy as sc
import seaborn as sns
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xchrom as xc
```

```python
path = '../'
rna4intra = sc.read_h5ad(f'{path}/0_data/processed_data/m_brain_paired_rna.h5ad') ## calculate ns,ls
atac4intra_raw = sc.read_h5ad(f'{path}/0_data/processed_data/m_brain_paired_atac.h5ad')  ## raw data
with h5py.File(f'{path}/1_XChrom/0_model_train/train_data//splits.h5', 'r') as hf:
    test_cellid = hf['test_cell'][:]
    trainval_peakid = hf['trainval_peak'][:]
atac4intra = atac4intra_raw[:,trainval_peakid] ## training data raw_ATAC
rna = rna4intra[test_cellid,] ## test data celltype label
save_path = './Results/'
```

```python
print(rna4intra.shape,atac4intra.shape,rna.shape)
```

```python
true = pd.read_feather(f'{save_path}/Seurat_true.feather') 
pred = pd.read_feather(f'{save_path}/Seurat_pred.feather') 
```

### 1- cross-cell prediciton test(auROC,auPRC)

```python
pred.index = pred.loc[:,'index']
pred = pred.drop('index',axis=1)
true.index = true.loc[:,'index']
true = true.drop('index',axis=1)
true = true.T
pred = pred.T
```

```python
true[true > 1] = 1
true_array = np.array(true,dtype=np.float32)
pred_array = np.array(pred,dtype=np.float32)
print(true_array.shape,pred_array.shape)
```

```python
## -1 Calculate overall auROC & auPRC 
overall_metrics = xc.tl.calc_auc_pr(true_array, pred_array, 'overall')
print(f"Overall auROC: {overall_metrics['auroc']:.4f}, auPRC: {overall_metrics['auprc']:.4f}")

## -2 Calculate per cell auRPC & auPRC
cell_metrics = xc.tl.calc_auc_pr(true_array, pred_array, 'percell')
print(f"Per-cell auROC: {cell_metrics['auroc']:.4f}, auPRC: {cell_metrics['auprc']:.4f}")
print(f"Valid cells: {cell_metrics['n_cells']}")

## -3 Calculate per peak auROC & auPRC
peak_metrics = xc.tl.calc_auc_pr(true_array, pred_array, 'perpeak')
print(f"Per-peak auROC: {peak_metrics['auroc']:.4f}, auPRC: {peak_metrics['auprc']:.4f}")
print(f"Valid peaks: {peak_metrics['n_peaks']}")
```

### 2- cross-cell impute test(ns,ls)

```python
assert atac4intra.obs.index.equals(rna4intra.obs.index), "Indexes differ!"
atac4intra.obs['pc32_leiden']=rna4intra.obs['pc32_leiden']
```

```python
# Rearrange the rows and columns of pred to ensure that the order is consistent with atac4intra
pred_sort = pred.reindex(index=rna.obs_names, columns=atac4intra.var_names)
```

```python
print(pred_sort.columns.equals(atac4intra.var_names))
```

```python
atac4intrap = atac4intra.copy()
atac4intrap.X[test_cellid, :] = pred_sort
```

```python
atac4intrap.write_h5ad(f'{save_path}/seurat_impALL.h5ad')
```

```python
# Seurat impute
ad1 = atac4intrap.copy()
ad1 = xc.tl.calc_pca(ad1)
ns,ls= xc.tl.calc_nsls_score(rna4intra,ad1,100,'pc32_leiden',test_cells =test_cellid,use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(100)={ns:.4f},label score(100)={ls:.4f}')

ns,ls= xc.tl.calc_nsls_score(rna4intra,ad1,50,'pc32_leiden',test_cells =test_cellid,use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(50)={ns:.4f},label score(50)={ls:.4f}')

ns,ls= xc.tl.calc_nsls_score(rna4intra,ad1,10,'pc32_leiden',test_cells =test_cellid,use_rep_rna = 'X_pca',use_rep_atac='X_pca')
print(f'neighbor score(10)={ns:.4f},label score(10)={ls:.4f}')

size_vector = [70 if idx in test_cellid else 20 for idx in range(ad1.n_obs)]
f, ax = plt.subplots(1, 1, figsize=(6, 4))
sc.pl.umap(ad1, color='pc32_leiden', size=size_vector, ax=ax, show=False)
```
