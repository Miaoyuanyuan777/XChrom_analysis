## Calculate the proportion of Mono subclusters(R1-R8) in the pbmc10k dataset
- Annotate the Mono subclusters of the PBMC10k dataset based on COVID-19 subclusters
- Calculate the proportion of Mono subclusters in the COVID-19 and PBMC10k datasets

```r
library(Seurat)
library(SeuratDisk)
library(ggplot2)
library(cowplot)
library(reticulate)
```

```r
path = '../../'
covidmono = paste0(path,'/0_preprocess/COVID19_MHH50/scRNA/Monocytes_scRNAseq.rds')
out_dir <- paste0(path,'/0_preprocess/pbmc10k/')
pbmc10kmono = paste0(path,'/0_preprocess/pbmc10k/raw_rna_mono.h5ad')
```

```r
ad <- reticulate::import("anndata")
adata <- ad$read_h5ad(pbmc10kmono)
adata$obs$index <- adata$obs_names
adata$var$index <- adata$var_names 
var_names <- as.character(adata$var$features)
obs_names <- as.character(adata$obs$index)
counts = adata$T$X
rownames(counts) <- var_names
colnames(counts) <- obs_names
cell_meta <- as.data.frame(adata$obs)
gene_meta <- as.data.frame(adata$var)
# create seurat object
seurat_obj <- Seurat::CreateSeuratObject(
  counts = counts,
  meta.data = cell_meta,
  assay = "RNA"
)
seurat_obj[["RNA"]]@meta.features <- gene_meta
```

```r
seurat_obj
# saveRDS(seurat_obj,paste0(out_dir,'pbmc10k_rna_mono.rds'))
```

```r
pbmc10k <- seurat_obj
covid <- readRDS(covidmono)
covid<-UpdateSeuratObject(covid)
p <- DimPlot(object = covid, reduction = "umap", group.by = "new.id", label = TRUE, label.size = 3, repel = TRUE)+ggtitle("COVID Mono subcluster")
print(p)
```

```r
DefaultAssay(pbmc10k) <- "RNA"
pbmc10k <- NormalizeData(object = pbmc10k, normalization.method = "LogNormalize") 
pbmc10k <- FindVariableFeatures(pbmc10k, nfeatures = 1000, selection.method = "vst", assay="RNA")
pbmc10k <- ScaleData(pbmc10k, features=VariableFeatures(pbmc10k))
covid <- RunUMAP(
  covid, 
  dims = 1:20, 
  reduction = "pca",
  reduction.name = "umap",
  return.model = TRUE
)
```

```r
anchors <- FindTransferAnchors(
  reference = covid,
  query = pbmc10k,
  # normalization.method = "SCT",
  reference.reduction = "pca", 
  dims = 1:20 
)

pbmc10k <- MapQuery(
  anchorset = anchors,
  query = pbmc10k,
  reference = covid,
  refdata = list(
    celltype.l1 = "celltypeL0",
    celltype.l2 = "new.id"
  ),
  reference.reduction = "pca",
  reduction.model = "umap"
)
```

```r
# saveRDS(pbmc10k,paste0(out_dir,'pbmc10k_mono_annoR1R8.rds'))
```

```r
pbmc10k@meta.data$new.id <- pbmc10k@meta.data$predicted.celltype.l2
table(pbmc10k@meta.data$new.id)
```

```r
reference_levels <- levels(covid$new.id)
pbmc10k$predicted.celltype.l2 <- factor(pbmc10k$predicted.celltype.l2, 
                                      levels = reference_levels)
ref_color_palette <- scales::hue_pal()(length(reference_levels))
names(ref_color_palette) <- reference_levels
```

```r
a1 <- subset(covid,subset = Severity == 'severe')
a2 <- subset(covid,subset = Severity == 'mild')
a3 <- subset(covid,subset = Severity == 'post')
p1<-DimPlot(object = a1, reduction = "umap", group.by = "new.id", label = TRUE, label.size = 3, repel = TRUE) +ggtitle('severe')#+ NoLegend()
p2<-DimPlot(object = a2, reduction = "umap", group.by = "new.id", label = TRUE, label.size = 3, repel = TRUE) +ggtitle('mild')#+ NoLegend()
p3<-DimPlot(object = a3, reduction = "umap", group.by = "new.id", label = TRUE, label.size = 3, repel = TRUE) +ggtitle('post')#+ NoLegend()
p4<-DimPlot(object = pbmc10k, reduction = "ref.umap", group.by = "predicted.celltype.l2", label = TRUE, label.size = 3, repel = TRUE) +
ggtitle('pbmc10k')+ scale_color_manual(values = ref_color_palette, breaks = reference_levels,drop = TRUE)
p5<-DimPlot(pbmc10k, reduction = "ref.umap", group.by = "predicted.celltype.l1", label = TRUE, label.size = 3, repel = TRUE)
p1
p2
p3
p4
p5
```

```r
get_proportion <- function(seurat_obj, group_name) {
  counts <- table(seurat_obj@meta.data$new.id)
  data.frame(
    R = names(counts),
    Count = as.numeric(counts),
    Proportion = as.numeric(prop.table(counts)),
    Group = group_name
  )
}
df_severe <- get_proportion(a1, "severe")
df_mild <- get_proportion(a2, "mild")
df_post <- get_proportion(a3, "post")
df_pbmc <- get_proportion(pbmc10k, "pbmc10k") 

combined_df <- rbind(df_severe, df_mild, df_post, df_pbmc)
```

```r
all_R <- paste0("R", 1:8)
combined_df$R <- factor(combined_df$R, levels = all_R)
combined_df$Group <- factor(combined_df$Group, 
                           levels = c("severe", "mild", "post",'pbmc10k'))
p6 <- ggplot(combined_df, aes(x = Group, y = Proportion ,fill = R)) + #, fill = Group
  geom_bar(stat = "identity", position = position_dodge(width = 0.8), width = 0.7) +
  facet_wrap(~ R, ncol = 8) +
  labs(
    title = "Proportion of Mono subclusters Across Conditions",
    x = "Condition",
    y = "Proportion",
    fill = "Condition"
  ) +
  scale_y_continuous(labels = scales::percent_format()) + 
  theme_bw(base_size = 12) +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    strip.background = element_blank(), 
    strip.text = element_text(size = 10, face = "bold"),
    legend.position = "bottom"
  )+NoLegend()
p6
```

```r
write.csv(combined_df,"./Proportion_MonoR1toR8.csv")
```
