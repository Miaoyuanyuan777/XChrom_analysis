## Remove the batches within the COVID-19 data and those with the PBMC10k data

```python
import scanpy as sc
import pandas as pd
import numpy as np
import anndata
import scanpy.external as sce
import seaborn as sns
import matplotlib.pyplot as plt
import xchrom as xc
```

```python
sc.set_figure_params(figsize=(3,3), frameon=False)
sns.set_theme()
%config InlineBackend.print_figure_kwargs={"facecolor": "w"}
%config InlineBackend.figure_format="retina"
```

```python
path = '../'
spath = f'{path}/0_preprocess/data/' 
rna = sc.read_h5ad(f'{spath}/mild_severe_all.h5ad')
print('Severity-cell numbers: \n',rna.obs.Severity.value_counts())
```

```python
rna_ms = anndata.AnnData(X=rna.raw.X, obs=rna.obs,var=rna.raw.var)
```

```python
rna_ms1 = xc.tl.calc_pca(rna_ms)
sc.pp.neighbors(rna_ms1)
sc.tl.umap(rna_ms1)
sc.pl.umap(rna_ms1, color=['batch','sample_assign'])
sc.pl.umap(rna_ms1, color=['celltypeL0','Severity'])
plt.tight_layout()
```

```python
pbmc10k = sc.read_h5ad(f'{path}/0_preprocess/pbmc10k/raw_data/pbmc10k_rna.h5ad')
pbmc10k.obs['batch'] = 'pbmc10k'
combined = pbmc10k.concatenate(rna_ms, join='inner', batch_key='dataset')
combined.raw = None
combined.write_h5ad(f'{spath}/harmony/0_bcd10k_beforeBatch.h5ad')
```

```python
adata = combined.copy()
adata = xc.tl.calc_pca(adata,max_value=10,n_comps=32)
adata.obs.index = adata.obs.index.str.replace(r'-\d+$', '', regex=True)
sc.pp.neighbors(adata, use_rep='X_pca')
sc.tl.umap(adata)
sc.pl.umap(adata, color=['batch','sample_assign'])
sc.pl.umap(adata, color=['dataset','celltypeL0'])
```

```python
adata1 = adata.copy()
sce.pp.harmony_integrate(adata1, 'batch',max_iter_harmony=30)
'X_pca_harmony' in adata1.obsm
```

```python
sc.pp.neighbors(adata1, use_rep='X_pca_harmony')
sc.tl.umap(adata1)
sc.pl.umap(adata1, color=['batch','sample_assign'])
sc.pl.umap(adata1, color=['dataset','celltypeL0'])
```

```python
adata1.obs.index = adata1.obs.index.str.replace(r'-\d+$', '', regex=True)
adata1.raw = None
adata1.write_h5ad(f'{spath}/harmony/1_bcd10k_harmonyPCA.h5ad')
```

```python
adata1 = sc.read_h5ad(f'{spath}/harmony/1_bcd10k_harmonyPCA.h5ad')
```

```python
s1 = adata1[adata1.obs['batch']=='pbmc10k']
s2 = adata1[adata1.obs['batch'].isin(['b', 'c','d']), :]
```

```python
s1.write_h5ad(f'{spath}/harmony/pbmc10k/0_harmonyPCA.h5ad')
s2.write_h5ad(f'{spath}/harmony/covid_all/0_harmonyPCA.h5ad')
```
